### UNIVERSAL FUNCTIONS (UFUNC)
Computation on NumPy arrays can be very fast, or it can be very slow. The key to
making it fast is to use vectorized operations, generally implemented through Num‐Py’s universal functions (ufuncs).
NumPy provides a convenient interface into statically typed, compiled routine. This is known as a vectorized operation.
You can accomplish this by simply performing an operation on the array, which will then be applied to each element.

```python
# Using Python's simple loop
def compute_reciprocals(values):
  result = np.empty(len(values))

  for i in range(len(values)):
    result[i] = 1.0/values[i]
  return result
```

```python
print('Processing array of size 1M using Python loop')
big_arr = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_arr)
```

```python
print('Processing array using NumPy.')
%timeit (1.0/big_arr)
```

## Pandas
Pandas is a newer package built on top of NumPy, and provides an
efficient implementation of a DataFrame. DataFrames are essentially multidimensional arrays with attached row and column labels, and often with heterogeneous types and/or missing data.

```python
import pandas as pd
pd.__version__
```

### PANDA'S OBJECTS
The three fundamental Pandas data structures:
* **Series**: A Pandas Series is a 1D array of indexed data. It can be created from a list or array using `pd.Series()`
* **DataFrame**: DataFrame is a 2D array and can be thought of either as a generalization of a NumPy array, or as a specialization of a Python dictionary.
* **Index**: It can be thought of either as an immutable array or as an
ordered set.

#### SERIES
Series object is basically interchangeable with a one-dimensional NumPy array. The essential difference is the presence of the index: <br/>
While the NumPy array has an *implicitly* defined integer index used
to access the values, the Pandas Series has an *explicitly* defined index associated with the values.


```python
series = pd.Series(['AAPL', 'ORCL', 'GOOG', 'MSFT', 'IBM'])
series
```

```python
print(f"Series value: {series.values}")
print(f"Series index: {series.index}")
```

```python
print('Using explicity Index')
series_1 = pd.Series(['AAPL', 'ORCL', 'GOOG', 'MSFT', 'IBM'], index=[2,4,6,8,10])
print(series_1)
print('Using explicity Non Integer Index')
series_2 = pd.Series(['AAPL', 'ORCL', 'GOOG', 'MSFT', 'IBM'], index=['a','b','c','d','e'])
print(series_2)
```

```python
print(f"Series value: {series_2.values}")
print(f"Series index: {series_2.index}")
```

Think of Series acting in many ways like a **one dimensional NumPy array**, and in many ways like a standard **Python dictionary**.

2. **Indexers: loc, iloc, and ix**: These slicing and indexing conventions can be a source of confusion.

**To avoid this confusion Pandas provides some special indexer attributes that explicitly expose certain indexing schemes.** </br>
* **loc**: allows indexing and slicing that always references the *explicit* index.
* **iloc**: allows indexing and slicing that always reference the *implicit* index

### HANDLING MISSING DATA
**Trade-Offs in Missing Data**<br/>
Two Strategies:
1. **Masking Approach**: The mask might be an entirely separate Boolean array, or it may involve appropriation of one bit in the data representation to locally indicate the null status of a value. **This adds Overhead in storage and computation.**
2. **Sentinel Approach**: the sentinel value could be some data-specific convention, such as indicating a missing integer value with `–9999 or some rare bit pattern`. **May require extra logic in CPU/GPU arithmetic.**

* **None:Pythanic missing data**: The first sentinel value used by Pandas is None, a Python singleton object that is often used for missing data in Python code. Because None is a Python object, it cannot be used in any arbitrary NumPy/Pandas array, but only in arrays with data type 'object' (i.e., arrays of Python objects).
* **NaN: Missing numerical data**: special floating-point value recognized by all systems that use the standard IEEE floating-point representation. Regardless of the operation, the result of arithmetic with NaN
will be another NaN

```python
# None and NaN
print("None and NaN\n")
vals_None = np.array([1, None, 3, 4])
vals_None
```

```python
vals_NaN = np.array([1, np.nan, 3, 4])

vals_NaN
```

**OPERATION ON NULL VALUES**
Pandas treats None and NaN as essentially interchangeable for indicating
missing or null values.
Here are some methods for detecting, removing and replacing null values in Pandas.
* isnull() -> Generate a Boolean mask indicating missing values
* notnull() -> Opposite of isnull().
* dropna() -> Return filtered version of data
* fillna() -> return a copy of data with missing values filled or imputed.

```python
states_data_imbalanced = pd.DataFrame([area, population]).T
states_data_imbalanced['code']=['AK', 'TX', 'CA', 'NY']
states_data_imbalanced
```

```python
print(f"notNull:\n{states_data_imbalanced.notnull()}")
```

```python
print("Drop Null Values")
print("We cannot drop single values from DataFrame. We can only drop full rows or full columns.\n")
```

```python
print("Drop Rows (default)\n")
states_data_imbalanced.dropna()
```

```python
print("Drop Columns (axis=1)\n")
states_data_imbalanced.dropna(axis=1)
```

```python
print("Drop Columns (axis=1) if all values are null\n")
states_data_imbalanced.dropna(axis=1, how='all')
```

```python
print("Fill missing values using various techniques.\n")
print("1. fillna(0)")
states_data_imbalanced.fillna(0)
```

```python
print("1. ffill. forward-fill to propagate the previous value forward")
states_data_imbalanced.ffill()
```

```python
print("1. bfill. backward-fill to propagate the next value backward")
states_data_imbalanced.bfill()
```

### Combining Datasets: Concat and Append
These operations can involve anything from very straightforward concatenation
of two different datasets, to more complicated database-style joins and merges
that correctly handle any overlaps between the datasets. Series and DataFrames are built with this type of operation in mind, and Pandas includes functions and methods that make this sort of data wrangling fast and straightforward.

```python
companies = [
    {'Ticker': 'AAPL', 'Company': 'Apple'},
    {'Ticker': 'GOOG', 'Company': 'Google'},
    {'Ticker': 'MSFT', 'Company': 'Microsoft'},
    {'Ticker': 'NVDA', 'Company': 'NVidia'},
    {'Ticker': 'AMZN', 'Company': 'Amazon'}
]
companies_data = pd.DataFrame(companies, columns=['Ticker', 'Company'])
companies_data
```

```python
companies_stock = pd.DataFrame([
    {'Ticker': 'AAPL', 'Price': 249.80},
    {'Ticker': 'GOOG', 'Price': 302.58},
    {'Ticker': 'MSFT', 'Price': 383.65},
    {'Ticker': 'NVDA', 'Price': 174.90},
    {'Ticker': 'AMZN', 'Price': 207.10}
])
companies_stock
```

**Concat operation**

```python
print("Simple Concatenation")
combined = pd.concat([companies_data, companies_stock], axis=1)
combined
```

**Merge Operation**

```python
print("Merge Dataframes")
merged = pd.merge(companies_data, companies_stock, on='Ticker')
merged
```
