# Visualizing Data
In this notebook we will explore different ways to visualize data. Data visualization helps to *explore and communicate* data.
There are variety of tools/libraries available. Here we will explore Matplotlib.

```python
# !pip3 install matplotlib
```

```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.30, 1075.9, 2865.5, 5979.6, 10289.7, 14958.30]
```

### Line Chart
These charts are good for showing trends

```python

plt.plot(years, gdp, color='green', marker='*',linestyle='solid')
plt.title('Nomnal GDP')
```

```python
variance = [2**n for n in range(10)]
bias_squared = variance[::-1]
total_error = [x+y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
# Make multiple calls to plt.plot
plt.plot(xs, variance, 'g-', label='variance') # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # red dot dashed line
plt.plot(xs, total_error, 'b:', label='total_error') # blue dotted line
plt.legend(loc=9) # This gives labels in top center
plt.xlabel("Model Complexity")
plt.xticks([])
```

### Bar Chart
Bar chart is a good choice when you want to show how some quantities variers among some **discrete** set of items.

```python
# Dataset
temperature = [40, 38, 65, 68, 69, 72, 75, 75, 65, 60, 55, 50]
months = ['Jan', 'Feb', 'March', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
```

```python
# Plot bars
plt.bar(range(len(months)), temperature)
plt.title('Mean Temperature')

# label y-axis
plt.ylabel('Celcius')
#label x-axis
plt.xlabel('Month')
plt.xticks(range(len(months)), months)
plt.show()
```

Bar Chart also helps with exploring how values are distributed. Let's explore.

```python
plt.bar(months, temperature)
plt.axis([0, 100, 0, 12])
plt.xticks([10 * i for i in range(11)])
plt.plot()
```

Watch out when setting Axis...
**When creating bar charts it is considered especially bad form for your y-axis not to start at 0, since this is an easy way to mislead people.**

```python
students = [105, 110]
years = [2025, 2026]

plt.bar(years, students, 0.8)
plt.xticks(years)
plt.axis([2024, 2027, 100, 115])
plt.title('Misleading')
plt.show()

plt.bar(years, students, 0.8)
plt.xticks(years)
plt.axis([2024, 2027, 0, 115]) # set offset
plt.title('Not Misleading')
plt.show()
```

### Scatterplot
It is a right choice for visualizing the relationship between two paired sets of data.

```python
# Relationship between number of friends your users have and the number of minutes they spend on the site every day
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['user1', 'user2', 'user3', 'user4', 'user5', 'user6', 'user7', 'user8', 'user9']
```

```python
plt.scatter(friends, minutes)
# label each point
for label, f_count, m_count in zip(labels, friends, minutes):
  plt.annotate(label,
               xy=(f_count, m_count), # put label with its point
               xytext=(-5, -15), #offset
               textcoords="offset points"
              )
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.plot()
```
